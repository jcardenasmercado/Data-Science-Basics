# Matplotlib basics
***


```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt # Importacion standard para matplotlib
import seaborn as sns # Importacion standard para seaborn

# Para que los plots se vean en el notebook (Solo es necesario usando jupyter notebook)
%matplotlib inline
```

### Plot b√°sico
***

```python
"""
    Los plots basicos con matplotlib se hacen de manera muy sencilla:
        plt.plot(x, y)
    A medida que avance el notebook se ir√° profundizando en la llamada a esta funcion seg√∫n lo que se necesite
"""

# Creando un plot basico (X vs Y)


x = np.linspace(0, 10, 21)
y = np.sqrt(x)

plt.plot(x,y)
plt.ylabel('Titulo para Y')
plt.xlabel('Titulo para X')
plt.title('Titulo del plot')
plt.show() # Para mostrar el plot
```

### Creando subplots
***

```python
"""
    Crear subplots en matplotlib es sencillo:
        plt.subplot(numero de filas, numero de columnas, numero de plot)
        ej: 2,2,1 -> el primer plot estar√° en 1,1
"""
# Plot en dos columnas
plt.subplot(1,2,1)
plt.plot(x,y,'b--') # 'b--' b:linea color azul --:el tipo de lineado; para mas info: https://matplotlib.org/stable/gallery/color/named_colors.html
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

```python
# Plot en 2x2
plt.subplot(2,2,1)
plt.plot(x,y,'b')
plt.subplot(2,2,2)
plt.plot(y,x,'r')
plt.subplot(2,2,3)
plt.plot(x*2,y*2,'b*-')
plt.subplot(2,2,4)
plt.plot(y*2,x*2,'r*-')
```

***
### Matplotlib orientado a objetos

```python
# Definiendo el objeto
fig = plt.figure()

# Ahora mismo nuestro objeto no es mas que un cuadro vacio por lo que le tenemos que agregar ejes
axes = fig.add_axes([0.1,0.1,0.8,0.8]) 

# Nuestra figura ya posee ejes por lo que podemos empezar a plotear en ella
axes.plot(x,y)
axes.set_xlabel('Variable X')
axes.set_ylabel('Variable Y')
axes.set_title('X vs Y')
```

```python
# Un plot dentro de otro plot

# Esto se puede hacer agregando mas ejes y ajustando su tama√±o

fig2 = plt.figure() # Definimos el objeto

axes1 = fig2.add_axes([0.1,0.1,0.8,0.8]) # Agregamos los ejes principales
axes2 = fig2.add_axes([0.2,0.53,0.4,0.3]) # Agregamos los ejes secundarios

# Agregamos los titulos respectivos y los datos
axes1.set_title('Y vs X')
axes1.set_xlabel('y')
axes1.set_ylabel('x')
axes1.plot(y,x,'g*-')

axes2.set_title('X vs Y')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.plot(x,y,'r--')
```

```python
"""
    plt.subplots(numero de filas, numero de columnas)
"""
# Subplots en matplotlib orientado a objetos

fig3, axes = plt.subplots(nrows=1, ncols=2) # se declara el objeto junto con los respectivos ejes

# para controlar los axes en un subplots hay que acceder como si fuese una lista
axes[0].set_title('Este es el primer plot')
axes[0].plot(x,y,'g*-')

axes[1].set_title('Este es el segundo plot')
axes[1].plot(y,x,'r*-')
```

```python
# Al tener un subplot con muchas filas y columnas, el proceso de ploteo se puede automatizar:

# subplot de dimension 1x5
fig4, axes = plt.subplots(nrows=1,ncols=5, figsize=(15,3))

i=0
for ax in axes:
    ax.set_title('Figura N¬∫'+str(i))
    ax.plot(x,y)
    i+=1
    
```

```python
# subplot de dimension 5x5
fig5, axes = plt.subplots(nrows=5,ncols=5, figsize=(14,17))

i=0
for row in axes:
    for ax in row:
        ax.set_title('figura N¬∫'+str(i))
        ax.plot(x,np.log(y*i+1),'o-.')
        i+=1
```

### Matplotlib nos permite personalizar nuestros plots a fondo, cambiando el tama√±o de la figura, la resoluci√≥n, el color de las lineas, estilos, etc

```python
# Cambiando el tama√±o y resolucion de un plot:

fig, axes = plt.subplots(figsize=(12,6), dpi=240)
axes.plot(x,y)
```

```python
# Tambien podemos agregar una leyenda a nuestra figura:

fig, axes = plt.subplots(figsize=(12,6), dpi=240)
axes.plot(x,y, label='X vs Y') # Agredando label=titulo podemos personalizar la legenda
axes.plot(y,x,'g', label='Y vs X')
axes.legend()

# Tambien podemos elegir donde posicionar la leyenda de nuestra figura:

axes.legend(loc=3) # esquina inferior izquierda
axes.legend(loc=4) # esquina inferior derecha
axes.legend(loc=2) # esquina superior izquierda
axes.legend(loc=1) # esquina superior derecha
```

```python
"""
    Matplotlib nos permite personalizar el color, el tipo de lineado y el punteado de nuestras figuras,
    para mas detalle visitar:
    
    -ESTILOS DE LINEA: https://matplotlib.org/3.0.3/gallery/lines_bars_and_markers/line_styles_reference.html
    
    -COLORES: https://matplotlib.org/stable/gallery/color/named_colors.html
        ademas de los colores disponibles por matplotlib, podemos agregar el que queramos en el formato
        de codigo RGB hexadecimal, ej: #ff0000 -> rojo
    
    -MARCADORES: https://matplotlib.org/stable/api/markers_api.html
    
    los argumentos a modificar de axis.plot() son los siguientes:
    -color: color
    
    -lw: ancho
    
    -alpha: transparencia
    
    -ls: estilo de lineado
    
    -marker: marcador (indica la interseccion punto a punto de x e y)
    
    -markersize: modifica el tama√±o del marcador
    
    -markerfacecolor: modifica el color interior del marcador
    
    -markeredgecolor: modifica el color del contorno del marcador    
"""

# Algunos ejemplos:
x = np.arange(-30,30,1)
y = 1/(1+(np.exp(-x)))

fig, ax = plt.subplots(nrows=3,ncols=3,figsize=(12,16), dpi=200)

ax[0][0].set_title('Linea Hex RGB')
ax[0][0].plot(x,y,color='#7ed63a')

ax[0][1].set_title('Linea con linestyle')
ax[0][1].plot(x,y,color='#7ed63a', ls=':')

ax[0][2].set_title('Linea con marcador')
ax[0][2].plot(x,y, color='#7ed63a', ls='-', marker='x')

ax[1][0].set_title('Linea con transparencia')
ax[1][0].plot(x,y, color='#3ad6d4', alpha=0.5)

ax[1][1].set_title('Linea con marcador de otro color')
ax[1][1].plot(x,y,color='#3ad6d4', marker='2', markeredgecolor='black')

ax[1][2].set_title('Linea mas ancha')
ax[1][2].plot(x,y,color='#3ad6d4', lw=3)

ax[2][0].set_title('Linea con marcador mas grande')
ax[2][0].plot(x,y,color='#96400b',lw=3, marker='*', markersize=8, markeredgecolor='#6c939e')

ax[2][1].set_title('Lineas con transparencia y marcadores')
ax[2][1].plot(x,y,color='#96400b', lw=3, marker='$A$', markersize=1, markeredgecolor='black', alpha=0.4)
ax[2][1].plot(y,x,color='#6c939e', lw=3, marker='$B$', markersize=1, markeredgecolor='yellow', alpha=0.4)

ax[2][2].set_title('Linea con marcador personalizado')
ax[2][2].plot(x,y,color='purple', lw=2, marker='s', ls='-', markersize=5, markerfacecolor='yellow', markeredgecolor='green' )


```
